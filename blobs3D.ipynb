```python
from sklearn.datasets import make_blobs
import numpy as np

n_clusters = 4
n_samples = 2000
random_state = 170
X, y = make_blobs(n_samples=n_samples, centers=[[-5,-5,-5],[0,0,0],[5,5,5],[10,10,10]], cluster_std = [2,2,2,2], n_features=2,random_state=random_state)
#X, y = make_blobs(n_samples=n_samples, centers=n_clusters, n_features=3,random_state=random_state)
```

```python
X.shape
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(X[:,0], X[:,1], X[:,2], c=y)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
